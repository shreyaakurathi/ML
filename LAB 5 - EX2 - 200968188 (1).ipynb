```python
#import libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
#drop rows and columns with misssing values
df.dropna(thresh=5, inplace=True)
```

```python
values = ['Id', 'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
df.drop(values, axis=1, inplace=True)
```

```python
#Split into X and Y
X = df.copy()
y = X.pop('SalePrice')
```

```python
#split into train and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#numerical df
n_c  = [cname for cname in X.columns if X[cname].dtype in ['int64', 'float64']]
df_nc= df[n_c]
df_nc
```

```python
#categorical df
c_c = [cname for cname in X.columns if X[cname].dtype == 'object']
df_cc = df[c_c]
df_cc
```

```python
#find unique values in categorical
cat_att = []
for i in df.columns:
    if df[i].dtype == 'O':
        print(i)
        cat_att.append(i)
        if df[i].nunique() <= 10:
            print(f"Number of unique values in column {i}:", df[i].nunique())
```

```python
#pipeline
# scaling the numerical columns
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='median')

numerical_transformer = Pipeline(steps = [
    ('scaler', RobustScaler()),
    ('imputer', num_imputer)
])

# one-hot encoding the categorical columns
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore')),
    ('imputer', cat_imputer)
])

# bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, n_c),
        ('cat', categorical_transformer, c_c),
      ])
```

```python
#lasso model
lasso = Lasso(alpha=0.1)

pipe_l = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', lasso)])
```

```python
pipe_l.fit(X_train, y_train)
```

```python
#accuracy score
pipe_l.score(X_test, y_test)
```

```python
#Ridge Model
ridge = Ridge(alpha=1.0)

pipe_r = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', ridge)])
```

```python
pipe_r.fit(X_train, y_train)
```

```python
#Accuracy score
pipe_r.score(X_test, y_test)
```
